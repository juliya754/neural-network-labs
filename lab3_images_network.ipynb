```
import glob 
from PIL import Image 
from pathlib import Path
import os

file_dir = Path('dataset5/A/*')
data_dict = {}
for directory in glob.glob(os.path.abspath(file_dir)):
    images = []
    for files in glob.glob(directory+'/color_*.png'):
        images.append(files)
    data_dict.setdefault(directory[-1], images)
import matplotlib.pyplot as plt
import cv2


f, ax = plt.subplots(4,6, figsize = (20,30))

for i , (k, v) in enumerate(data_dict.items()):
  if k=='b':
    print(i, k, v)
    for j in range(0,24):
      img = cv2.imread(v[j])
      img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
      img = cv2.resize(img, (64,64))
      ax[j//6, j%6].imshow(img)
      ax[j//6, j%6].axis('off')

plt.show()
```

Объединение картинок, полученных от разных людей (директории A, B, C, D, Е)

```
!rsync -avp --remove-source-files 'dataset5/B/' 'dataset5/A/'
!rsync -avp --remove-source-files 'dataset5/C/' 'dataset5/A/'
!rsync -avp --remove-source-files 'dataset5/D/' 'dataset5/A/'
!rsync -avp --remove-source-files 'dataset5/E/' 'dataset5/A/'
```

```
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1. / 255,
    shear_range=0.2,
    zoom_range=0.2,
    validation_split=0.2)
```

```
data_dir = 'dataset5/A/'
image_size = 128
batch_size=32
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(image_size, image_size),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True,
    color_mode='rgb',
    subset='training')
```

```
test_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(image_size, image_size),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True,
    color_mode='rgb',
    subset='validation')
```

```
x, y = test_generator[0]
```

```
print(x.shape)
```

Построение и компиляция нейронной сети

```
from keras.layers import Conv2D, BatchNormalization, MaxPooling2D, Dense, Flatten, Dropout, Activation
from keras.models import Sequential
model = Sequential()
model.add(Conv2D(32, (3,3), input_shape = (128, 128, 3), activation = 'relu')) # сверточный слой: 32 фильтра размером 3х3, входное изображение 128х128 с 3 каналами (цветное)
model.add(BatchNormalization()) # слой нормализации (повышение точности, борьба с взрывающими и угасающими градиентами)
model.add(MaxPooling2D(2,2)) # подвыборка по максимальному значению: матрица 2х2, снижение размерности
model.add(Dropout(0.25)) 

model.add(Conv2D(64, (3,3), activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.25))

model.add(Conv2D(128, (3,3), activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.25))

model.add(Flatten()) # преобразование в одномерный вектор
model.add(Dense(512, activation = 'relu')) # полносвязный слой: 512 нейронов
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(24, activation = 'softmax')) # 24 класса на выходе

model.compile(loss = 'categorical_crossentropy', optimizer = 'rmsprop', metrics = ['accuracy']) # categorical_crossentropy - вычисляет потери между истинными и предсказанными метками
model.summary()
```

Обучение нейронной сети

```
history = model.fit_generator(
      train_generator,
      steps_per_epoch=train_generator.samples // batch_size,
      epochs=3)
```

```
predictions = model.predict(x)
```

```
model.evaluate(test_generator)
```

```
import numpy as np
print("predictions:", np.argmax(predictions, axis=-1))
print("real:", np.argmax(y, axis=-1))
```
